```python
def make_beta_pdf(success, failure, title = None, 
        plot_width= 250, plot_height = 250):
    p = figure(title = title, plot_height = plot_height, plot_width = plot_width)
    ns = np.arange(0,1, .001)
    ps = beta.pdf(ns, success, failure)
    p.line(x =ns, y = ps)
    return p
```

```python
"""From Bayesian statistics the fun way
2 click, 3 do not

"""
def make_posterior(prior, liklihood, plot_width = 300, plot_height = 300):
    p_liklihood = make_beta_pdf(liklihood[0], liklihood[1], 
        title = 'Liklihood', plot_height = plot_height, plot_width= plot_width)
    p_prior =  make_beta_pdf(prior[0], prior[1], title = 'Prior',
        plot_height = plot_height, plot_width = plot_width)
    p_post = make_beta_pdf(prior[0] + liklihood[0], prior[1] + liklihood[1], 
        title = 'Posterior', plot_width = plot_width, plot_height = plot_height)
    grid = gridplot([p_liklihood, p_prior, p_post], ncols = 3)
    return grid
prior = (5, 200)
liklihood = (86, 300 - 86)
liklihood = (2, 3)
show(make_posterior(prior, liklihood))
```

```python
f = 100
prior = (3, 7)
prior = (f * prior[0], f * prior[1])
liklihood = (36, 114)
show(make_posterior(prior, liklihood))
```
