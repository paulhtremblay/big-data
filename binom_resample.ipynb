```python
def resample(l):
    final = []
    for i in range(len(l)):
        final.append(random.choice(l))
    return final

def repeat_resample(sample, num_iter = 1000):
    means = []
    for i in range(num_iter):
        resample_s = resample(sample)
        means.append(np.mean(resample_s) )
    return means
```
