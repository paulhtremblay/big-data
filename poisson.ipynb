```python
def make_binom_graph(p = None, p_value = .05, n = 1000):
    if not p:
        p =  figure()
    y = []
    for k in range(1000):
        y.append(binom.pmf(n=1000, k =k, p =p_value))
    p.line(x = list(range(1000)), y = y)
    return p
show(make_binom_graph())
```

```python
def make_poisson_graph(p = None, p_value = .05, n = 1000):
    if not p:
        p =  figure()
    y = []
    for k in range(n):
        y.append(poisson.pmf(k, p_value * n, loc=0))
    x = list(range(n))
    p.line(x = x, y = y)
    return p
show(make_poisson_graph(p_value = .05))
```

```python
def make_both_graph(p = None, p_value = .05, n = 1000):
    if not p:
        p =  figure()
    y1 = []
    y2 = []
    print(p_value * n)
    for k in range(n):
        y1.append(poisson.pmf(k, p_value * n, loc=0))
        y2.append(binom.pmf(n=n, k =k, p =p_value))
    x = list(range(n))
    p.line(x = x, y = y1, legend_label = 'poission')
    p.line(x = x, y = y2, legend_label = 'binom', color= 'red')
    return p
show(make_both_graph(n=500, p_value = .5))
```

```python
def covid_rates(date, num_infected):
    #example of covid
    final = {}
    for i  in range(14):
        p = poisson.pmf(i, 5)
        final[date] = p * num_infected
        date -= datetime.timedelta(days = 1)
    return final
    

    
    #show(make_both_graph(n=14, p_value = 5/14))
    
def mult_days_covid():
    def get_x_y(data):
        x = sorted(data.keys())
        y = [data[key] for key in x]
        return x,y
        
    day1 = covid_rates(date = datetime.datetime(2020, 9, 1), num_infected = 500)
    day2 = covid_rates(date = datetime.datetime(2020, 9, 2), num_infected = 450)
    day3 = covid_rates(date = datetime.datetime(2020, 9, 3), num_infected = 400)

    p = figure(x_axis_type = 'datetime')
    p.line(x = get_x_y(day1)[0], y = get_x_y(day1)[1], legend_label = 'day1')
    p.line(x = get_x_y(day2)[0], y = get_x_y(day2)[1], legend_label = 'day2', color ='red')
    p.line(x = get_x_y(day3)[0], y = get_x_y(day3)[1], legend_label = 'day3', color ='green')


    show(p)

mult_days_covid()
```
